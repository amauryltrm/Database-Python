# Utilisation de la bibliothèque Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Charger un dataset intégré de seaborn
df = sns.load_dataset("penguins")

# Aperçu des données
print(df.head())

# Infos sur le dataset
print(df.info())
```

## Scatterplot

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species", style="sex")
plt.title("Relation longueur vs profondeur du bec (penguins)")
plt.show()
```

## Lineplot

```python
# Exemple avec un dataset temporel intégré
flights = sns.load_dataset("flights")

plt.figure(figsize=(7,4))
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
plt.title("Évolution du nombre de passagers par mois")
plt.show()
```

## Histogramme et KDE

```python
plt.figure(figsize=(6,4))
sns.histplot(data=df, x="flipper_length_mm", bins=20, kde=True, hue="species")
plt.title("Distribution de la longueur des nageoires")
plt.show()
```

## Barplot et Countplot

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="species", y="body_mass_g", ci="sd")
plt.title("Poids moyen par espèce (barplot)")
plt.show()

plt.figure(figsize=(6,4))
sns.countplot(data=df, x="island", hue="species")
plt.title("Nombre de manchots par île et par espèce (countplot)")
plt.show()
```

## Boxplot et Violinplot

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="species", y="body_mass_g", hue="sex")
plt.title("Boxplot du poids par espèce et sexe")
plt.show()

plt.figure(figsize=(6,4))
sns.violinplot(data=df, x="species", y="flipper_length_mm", hue="sex", split=True)
plt.title("Violinplot longueur des nageoires par espèce et sexe")
plt.show()
```

## Stripplot et Swarmplot

```python
plt.figure(figsize=(6,4))
sns.stripplot(data=df, x="species", y="bill_length_mm", jitter=True, hue="sex", dodge=True)
plt.title("Stripplot : Longueur du bec")
plt.show()

plt.figure(figsize=(6,4))
sns.swarmplot(data=df, x="species", y="bill_length_mm", hue="sex", dodge=True)
plt.title("Swarmplot : Longueur du bec")
plt.show()
```

## Heatmap (corrélations)

```python
plt.figure(figsize=(6,4))
corr = df.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matrice de corrélation")
plt.show()
```

## Pairplot

```python
sns.pairplot(df.dropna(), hue="species")
plt.suptitle("Pairplot des variables numériques par espèce", y=1.02)
plt.show()
```

## Styles (darkgrid, white, ticks, …)

```python
for style in ["darkgrid", "whitegrid", "dark", "white", "ticks"]:
    sns.set_style(style)
    plt.figure(figsize=(5,3))
    sns.histplot(data=df, x="flipper_length_mm", bins=15, kde=True)
    plt.title(f"Style = {style}")
    plt.show()

# Remettre un style par défaut
sns.set_style("darkgrid")
```

## Palettes de couleurs

```python
palettes = ["deep", "muted", "bright", "pastel", "dark", "colorblind"]

for pal in palettes:
    plt.figure(figsize=(5,3))
    sns.barplot(data=df, x="species", y="body_mass_g", palette=pal)
    plt.title(f"Palette = {pal}")
    plt.show()
```

## Faceting (FacetGrid et catplot)

```python
# Exemple 1 : FacetGrid
g = sns.FacetGrid(df, col="species", hue="sex", height=4)
g.map_dataframe(sns.scatterplot, x="bill_length_mm", y="bill_depth_mm")
g.add_legend()
plt.show()

# Exemple 2 : catplot
sns.catplot(data=df, x="species", y="flipper_length_mm", hue="sex", kind="violin", col="island")
plt.show()
```

## Cheat Sheet Seaborn

```python
# ===============================
# 1. Graphiques de base
# ===============================
# Scatterplot
sns.scatterplot(data=df, x="bill_length_mm", y="bill_depth_mm", hue="species")

# Lineplot
sns.lineplot(data=sns.load_dataset("flights"), x="year", y="passengers", hue="month")

# Histogramme + KDE
sns.histplot(data=df, x="flipper_length_mm", bins=20, kde=True, hue="species")

# Barplot (valeurs moyennes)
sns.barplot(data=df, x="species", y="body_mass_g", ci="sd")

# Countplot (fréquences)
sns.countplot(data=df, x="island", hue="species")

# ===============================
# 2. Graphiques avancés
# ===============================
# Boxplot
sns.boxplot(data=df, x="species", y="body_mass_g", hue="sex")

# Violinplot
sns.violinplot(data=df, x="species", y="flipper_length_mm", hue="sex", split=True)

# Stripplot
sns.stripplot(data=df, x="species", y="bill_length_mm", hue="sex", dodge=True)

# Swarmplot
sns.swarmplot(data=df, x="species", y="bill_length_mm", hue="sex", dodge=True)

# Heatmap des corrélations
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

# Pairplot
sns.pairplot(df.dropna(), hue="species")

# ===============================
# 3. Styles et palettes
# ===============================
sns.set_style("darkgrid")
sns.set_palette("muted")

# ===============================
# 4. Faceting
# ===============================
sns.catplot(data=df, x="species", y="flipper_length_mm", hue="sex", kind="violin", col="island")
```
